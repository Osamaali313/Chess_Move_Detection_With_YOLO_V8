```python
import ultralytics

from ultralytics import YOLO
import os
from PIL import Image
import cv2
from IPython.display import Video
import glob
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')


ultralytics.checks()
```

```python
model = YOLO('yolov8s.pt')
```

```python
Root_dir = '/content/chess-pieces-detection-image-dataset/Chess_pieces'
```

```python
yaml_path = os.path.join(Root_dir, 'data.yaml.yaml')

train_path = os.path.join(Root_dir, 'train', 'images')
valid_path = os.path.join(Root_dir, 'valid', 'images')
```

```python
results = model.train(
    data=yaml_path,
    epochs=15,
    batch=32,
    lr0=0.0001,
    lrf=0.1,
    imgsz=640,
    plots= True
)
```

```python
data_dir = "/content/runs/detect/train3"
img_dir = os.path.join(data_dir, '*g')
files = glob.glob(img_dir)
imgs = []
for image in files:
    img = cv2.imread(image)
    imgs.append(img)
    plt.figure(figsize=(10,10), dpi= 200)
    plt.imshow(img)
```

```python
image_test_path = '/content/chess-pieces-detection-image-dataset/Chess_pieces/test/7a34d8620235048917b28bcfd3b5572b_jpg.rf.450c577e3be66b5232c54ffc9ec9e6b7.jpg'

results = model.predict(source=image_test_path,
                        imgsz=640)

test_image = results[0].plot(line_width=2)
plt.imshow(test_image)
```
